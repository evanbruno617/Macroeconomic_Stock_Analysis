```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#read in inflation data
inflation = pd.read_excel("Inflation/SeriesReport-20220623191557_2affe5.xlsx")
inflation.head()
```

```python
# drop the first 10 rows
inflation.drop(inflation.index[0:10], inplace= True)
```

```python
# change row 10 to the column names
inflation.columns = inflation.loc[inflation.index==10].values[0].tolist()
```

```python
#get ride of uncessary columns
inflation.drop(index = 10, columns=["HALF1", "HALF2"], inplace = True)
inflation.head()
```

```python
#switch out the column month names with integers
new_columns = ["Year"]
for i in inflation.columns.tolist():
    if i == "Jan":
        new_columns.append("1")
    if i == "Feb":
        new_columns.append('2')
    if i == "Mar":
        new_columns.append('3')
    if i == "Apr":
        new_columns.append('4')
    if i == "May":
        new_columns.append('5')
    if i == "Jun":
        new_columns.append('6')
    if i == "Jul":
        new_columns.append('7')
    if i == "Aug":
        new_columns.append('8')
    if i == "Sep":
        new_columns.append('9')
    if i == "Oct":
        new_columns.append('10')
    if i == "Nov":
        new_columns.append('11')
    if i == "Dec":
        new_columns.append('12')
```

```python
#replace columns with new columns
inflation.columns = new_columns
inflation.head()
```

```python
#change the year type to str
inflation["Year"] = inflation["Year"].astype(str)
```

```python
#to CSV
inflation.to_csv("Clean_Data/inflation.csv")
```
