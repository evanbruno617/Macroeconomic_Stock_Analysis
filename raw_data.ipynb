```python
def cleaning(data):
    #filter after 2002
    data = data.loc[data["DATE"] >= "2002-01-01"]

    #create new column with only the year
    year = []
    for i in data["DATE"]:
        year.append(i[0:4])
    data["Year"] = year

    # create new column with only the month
    Date = []
    for i in data["DATE"]:
        if i[5] == "0":
            Date.append(i[6])
        else:
            Date.append(i[5:7])
    data["Month"] = Date
    return data
```

```python
# industrial output 
indust = pd.read_csv("RawData/INDPRO.csv")
indust.head()
```

```python
# call function
indust_df = cleaning(indust)
indust_df.head()
```

```python
#to CSV
indust_df.to_csv("Clean_Data/Industrial.csv", index=False)
```
